```python
with open("./Dataset/oos/data_full.json") as f:
    data = json.load(f)
    
oos_train = data.get('oos_train')
oos_val = data.get('oos_val')
val = data.get('val')
train = data.get('train')

train_data = train + oos_train
test_data = val + oos_val
```

```python
tmp_labels = []
tmp_labels.append('oos')
for s, l in train_data:
    tmp_labels.append(l)
tmp_labels = list(set(tmp_labels))
tmp_num = [i for i in range(len(tmp_labels))]
label_dict = dict(zip(tmp_labels, tmp_num))
```

```python
import re
intents = []
for l in label_dict:
    template = json.loads(json.dumps({"intents" : intent}))
    template["intents"]["name"] = l
    for t in train_data:
        if l == t[1]:
            tt = re.sub('[0-9:;\-=+(){}/|!?@#$%^&*â€™\'\"?,.]', '', t[0])
            if not any(tt in i for i in template["intents"]["sampleUtterances"]):
                template["intents"]["sampleUtterances"].append(tt)
    intents.append(template["intents"])
```
